```python
#if you need an API Key from OpenAI
#https://platform.openai.com/account/api-keys
# pip install openai
# pip install python-dotenv

from openai import OpenAI
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())

OPENAI_API_KEY  = os.getenv('OPENAI_API_KEY')
```

```python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)

def continue_conversation(messages, temperature=0):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=temperature,
    )
    #print(str(response.choices[0].message["content"]))
    return response.choices[0].message.content
```

```python
def add_prompts_conversation(_, context):
    #Get the value introduced by the user
    prompt = client_prompt.value_input
    client_prompt.value = ''

    #Append to the context the User prompt.
    context.append({'role':'user', 'content':f"{prompt}"})

    #Get the response.
    response = continue_conversation(context)

    #Add the response to the context.
    context.append({'role':'assistant', 'content':f"{response}"})

    #Update the panels to show the conversation.
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(
        pn.Row('Assistant:', pn.pane.Markdown(response, width=600)))

    return pn.Column(*panels)
```

# One-Page Report — Prompt Experiments with GPT 3.5

## Overview
I conducted experiments with three different chatbot prompts using GPT-3.5 in a Panel-based interactive notebook. Each prompt represented a distinct domain: a **Song Library**, a **Doctor’s Secretary**, and a **Travel Assistant**. The goal was to explore how prompt variations influence GPT’s behavior, accuracy, and user experience.

---

## Prompt 1 — Song Library
- **Behavior Observed:** GPT correctly greeted the user, asked for a song title, and responded with a matching song. It also generated a “play button” placeholder.
- **Strengths:** Friendly, concise, clear role understanding, smooth conversation flow.
- **Limitations:** Cannot actually play music or render interactive buttons.
- **Hallucinations / Errors:** Minor - for uncommon song titles, GPT sometimes suggested a song by a random or nonexistent artist.

---

## Prompt 2 — Doctor Secretary
- **Behavior Observed:** GPT managed patient scheduling, understood doctor’s availability, and denied appointments on unavailable days. Handled emergencies appropriately.
- **Strengths:** Role-playing was clear, polite, context-aware, confirmed details before booking.
- **Limitations:** Occasionally contradicted itself regarding unavailable days and suggested appointments inconsistently at first.
- **Hallucinations / Errors:** Serious - initially suggested impossible dates before correcting itself; shows GPT can misinterpret complex availability rules since it forgets previous conversation.

---

## Prompt 3 — Travel Assistant
- **Behavior Observed:** GPT asked for travel details, provided flight options with times and approximate prices, and allowed user selection. Adjusted style when instructed to be concise.
- **Strengths:** Clear domain understanding, structured options, adapted style based on user input.
- **Limitations:** Responses sometimes verbose; needed user reminders to be brief.
- **Hallucinations / Errors:** Minor - flight times and prices are illustrative, not real-time accurate.

---

## General Observations
1. **Prompt specificity is critical.** Detailed system prompts improve accuracy and context-awareness.
2. **Short vs. long responses:** GPT tends to be verbose unless explicitly instructed.
3. **Domain accuracy:** Handles structured tasks well but can hallucinate factual data (song availability, flight schedules).
4. **Handling edge cases:** GPT generally responds politely to emergencies; effectiveness depends on prompt clarity.

---

## Conclusions / Learnings
- GPT behavior depends heavily on prompt design: defining roles, rules, and constraints improves performance.
- Testing multiple prompt variations helps identify hallucinations, errors, or verbosity issues.
- Interactive GPT-based chatbots are suitable for prototyping role-based assistants but cannot perform real-world actions (e.g., play music, book flights).
- Balancing friendliness, accuracy, and conciseness is essential for practical chatbot applications.
